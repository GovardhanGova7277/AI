Q1. Extract Clean Email Usernames

You receive a list of messy email IDs. Using list comprehension, extract only the username part.

```python
emails = ["john.doe@gmail.com ", "  ana123@yahoo.com", "mark_z@outlook.com"]

names = [name[:name.index('@')].strip() for name in emails]

names
```

Q2. Product Prices With 18% GST

Given a list of base prices, return prices including GST.

```python
prices = [100, 250, 399, 1200]

gst_prices = [price+18//100 for price in prices]

gst_prices
```

Q3. Square Only Even Numbers From Mixed List

```python
nums = [10, 15, 20, 21, 22]

sq_even_nums = [num**2 for num in nums if num % 2 == 0]

sq_even_nums
```

Q1. Reverse a Number Without Using Slicing

```python
n = 12345
res=0
while n>0:
    res = res*10 + n%10
    n = n // 10
print(res)
```

Q2. Count Vowels in a Sentence

Ignore case.


```python
sentence = "Today I have build a agent"
result=0
for i in sentence:
    if i.lower() in ('a','e','i','o','u'):
        result+=1
print(f"Number of vowels in the sentence {result}")
```

Q4. Remove Duplicates Without Using Set

Maintain order.

```python
li = [3,6,7,3,8,6]
seen={}
out = []
for num in li:
    if num not in seen:
        seen[num] = True
        out.append(num)
print(out)
```
