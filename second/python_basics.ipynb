$$ Print area of triangle given sides $$
$$ semi perimeter : s = 0.5(a+b+c) $$
$$ area : (s(s-a)(s-b)(s-c))**0.5

```python
a=float(input("Enter a value: "))
b=float(input("Enter b value: "))
c=float(input("Enter c value: "))
```

```python
print(a,b,c)
```

```python
s=0.5*(a+b+c)
area = (s*(s-a)*(s-b)*(s-c))**0.5
```

```python
print("The area of : ",area)
```

```python
# 6.	Python Program to Print all Prime Numbers between 1 and 100

for num in range(2,101):
    is_prime = True

    for i in range(2, int(num**0.5)+1):
        if num % i == 0:
            is_prime = False
            break
    if is_prime:
        print(num,end=" ")
```
